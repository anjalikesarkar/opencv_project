# HISTOGRAM

# Gray Image Hist

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('../images/smily_women.jpg')
img = cv2.resize(img,(500,400))
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#channel = 0 for gray image
hist = cv2.calcHist([img],channels = [0], mask = None,histSize=[256], ranges=[0,256] )


plt.figure(figsize=(4, 4))  # Width = 8 inches, Height = 6 inches
plt.plot(hist)
plt.title('Grayscale Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()

cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Color Image Hist

```python
img = cv2.imread('../images/car.jpg')
img = cv2.resize(img,(400,300))

plt.figure(figsize=(6, 4))

color= ('b','g','r')
for i, col in enumerate(color):
    hist = cv2.calcHist([img],channels =[i],mask = None,histSize=[256],ranges=[0,256])
    plt.plot(hist,color=col)


plt.title('Color Histogram')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 256])
plt.show()

cv2.imshow('ColoHist Image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# HISTOGRAM EQUALIZATION

```python
img = cv2.imread('../images/blur.jpg')
org_img = cv2.resize(img, (400,300))
org_img = cv2.cvtColor(org_img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(8,4))
#check 1st histo on original image
org_hist = cv2.calcHist([org_img], channels = [0],mask = None, histSize=[256],ranges=[0,256])
plt.subplot(1,2,1)
plt.plot(org_hist)
plt.title('Original Image Hist')
plt.xlim([0, 256]) 

eq_img = cv2.equalizeHist(org_img)
eq_hist = cv2.calcHist([eq_img], channels = [0],mask = None, histSize=[256],ranges=[0,256])
plt.subplot(1,2,2)
plt.plot(eq_hist)
plt.title('Equalized Image Hist')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 256])  


cv2.imshow('Equalized_image',eq_img)
cv2.imshow('Original_Image',org_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.show()
```

# ( CLAHE )Contrast Limited Adaptive Histogram Equalization

```python
img = cv2.imread('../images/blur.jpg')

org_img = cv2.resize(img, (400,300))
org_img = cv2.cvtColor(org_img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(8,4))
#check 1st histo on original image
org_hist = cv2.calcHist([org_img], channels = [0],mask = None, histSize=[256],ranges=[0,256])
plt.subplot(1,2,1)
plt.plot(org_hist)
plt.title('Original Image Hist')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 256]) 

cl = cv2.createCLAHE(clipLimit = 1,tileGridSize=(8,8))

cl_img = cl.apply(org_img)

cl_hist = cv2.calcHist([cl_img], channels = [0],mask = None, histSize=[256],ranges=[0,256])
plt.subplot(1,2,2)
plt.plot(cl_hist)
plt.title('CLACH Image Hist')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 256])

cv2.imshow('CLACH_Equalized_image',cl_img)
cv2.imshow('Original_Image',org_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.show()
```

# ORIGIVAL ,EQUALIZER,CLAHE EQUALIZER

```python
img = cv2.imread('../images/blur2.png')

org_img = cv2.resize(img, (400,300))
org_img = cv2.cvtColor(org_img, cv2.COLOR_BGR2GRAY)
org_hist = cv2.calcHist([org_img], channels = [0],mask = None, histSize=[256],ranges=[0,256])

eq_img = cv2.equalizeHist(org_img)
eq_hist = cv2.calcHist([eq_img], channels = [0],mask = None, histSize=[256],ranges=[0,256])

cl = cv2.createCLAHE(clipLimit = 1,tileGridSize=(8,8))
cl_img = cl.apply(org_img)
cl_hist = cv2.calcHist([cl_img], channels = [0],mask = None, histSize=[256],ranges=[0,256])

plt.figure(figsize = (6,4))
plt.plot(org_hist, color = 'black')
plt.plot(eq_hist, color='red')
plt.plot(cl_hist, color='green')

plt.title('Image Hist')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.xlim([0, 256])

cv2.imshow('Equalized_image',eq_img)
cv2.imshow('CLACH_Equalized_image',cl_img)
cv2.imshow('Original_Image',org_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.show()
```
